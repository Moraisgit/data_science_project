We've already perceived that data is recorded yearly, but we can try other aggregations...

```python
from pandas import Index, Period, Series, DataFrame
import pandas as pd
import numpy as np

def ts_aggregation_by(
    data: Series | DataFrame,
    gran_level: str = "D",
    agg_func: str = "mean",
    custom_years: int = None  # New argument for multi-year aggregation
) -> Series | DataFrame:
    df: Series | DataFrame = data.copy()
    
    if custom_years:  
        year_index = df.index.year
        custom_period = (year_index // custom_years) * custom_years
        df.index = pd.to_datetime(custom_period, format="%Y")
        df.index.name = f"{custom_years}_year_period"
    else:
        # Default time granularity
        index: Index[Period] = df.index.to_period(gran_level)
        df.index = index.to_timestamp()
    
    # Group by the new index and aggregate
    df = df.groupby(df.index, dropna=True, sort=True).agg(agg_func)
    
    return df

ss_second_years: Series = ts_aggregation_by(series, custom_years=SECOND)

figure()
plot_line_chart(
    ss_second_years.index.to_list(),
    ss_second_years.to_list(),
    xlabel="Year",
    ylabel=target,
    title=f"{file_tag} {SECOND}-yearly {target}",
)
savefig(f"{path_to_directory}/images/Set 2 - Second granularity.png", bbox_inches='tight')
show()

# # Plot the results
# import matplotlib.pyplot as plt

# plt.figure()
# plt.plot(ss_3_years.index, ss_3_years, marker='o', linestyle='-')
# plt.xlabel("3-Year Periods")
# plt.ylabel("Mean Value")
# plt.title("3-Year Aggregated Time Series")
# plt.grid()
# plt.show()
```

### Autocorrelation

```python
from dslabs_functions import plot_multiline_chart


def get_lagged_series(series: Series, max_lag: int, delta: int = 1):
    lagged_series: dict = {"original": series, "lag 1": series.shift(1)}
    for i in range(delta, max_lag + 1, delta):
        lagged_series[f"lag {i}"] = series.shift(i)
    return lagged_series


# figure(figsize=(3 * HEIGHT, HEIGHT))
figure()
lags = get_lagged_series(series, 20, 10)
plot_multiline_chart(series.index.to_list(), lags, xlabel=index, ylabel=target)
savefig(f"{path_to_directory}/images/Set 2 - Autocorrelation: lag-plots.png", bbox_inches='tight')
```

```python
from matplotlib.pyplot import setp
from matplotlib.gridspec import GridSpec


def autocorrelation_study(series: Series, max_lag: int, delta: int = 1):
    k: int = int(max_lag / delta)
    fig = figure(figsize=(4 * HEIGHT, 2 * HEIGHT), constrained_layout=True)
    gs = GridSpec(2, k, figure=fig)

    series_values: list = series.tolist()
    for i in range(1, k + 1):
        ax = fig.add_subplot(gs[0, i - 1])
        lag = i * delta
        ax.scatter(series.shift(lag).tolist(), series_values)
        ax.set_xlabel(f"lag {lag}")
        ax.set_ylabel("original")
    ax = fig.add_subplot(gs[1, :])
    ax.acorr(series, maxlags=max_lag)
    ax.set_title("Autocorrelation")
    ax.set_xlabel("Lags")
    return


autocorrelation_study(series, 10, 1)
savefig(f"{path_to_directory}/images/Set 2 - Autocorrelation: correlogram.png", bbox_inches='tight')
```

## **Data Stationarity**
### Seasonality

```python
from pandas import Series
from matplotlib.pyplot import subplots, show, gca
from matplotlib.figure import Figure
from matplotlib.axes import Axes
from statsmodels.tsa.seasonal import DecomposeResult, seasonal_decompose
from dslabs_functions import HEIGHT, set_chart_labels

from pandas import DataFrame, Series, read_csv
from matplotlib.pyplot import figure, show
from dslabs_functions import plot_line_chart

def plot_components(
    series: Series,
    title: str = "",
    x_label: str = "time",
    y_label: str = "",
) -> list[Axes]:
    decomposition: DecomposeResult = seasonal_decompose(series, model="add")
    components: dict = {
        "observed": series,
        "trend": decomposition.trend,
        "seasonal": decomposition.seasonal,
        "residual": decomposition.resid,
    }
    rows: int = len(components)
    fig: Figure
    axs: list[Axes]
    fig, axs = subplots(rows, 1, figsize=(4, rows * HEIGHT))
    fig.suptitle(f"{title}")
    i: int = 0
    for key in components:
        set_chart_labels(axs[i], title=key, xlabel=x_label, ylabel=y_label)
        axs[i].plot(components[key])
        i += 1
    return axs


# file_tag = "ASHRAE"
# filename = "data/time_series/ashrae.csv"
# index = "timestamp"
# target = "meter_reading"
data: DataFrame = read_csv(
    filename,
    index_col=index,
    sep=",",
    decimal=".",
    parse_dates=True,
    infer_datetime_format=True,
)
series: Series = data[target]

plot_components(
    series,
    title=f"{file_tag} Yearly {target}",
    x_label=series.index.name,
    y_label=target,
)

savefig(f"{path_to_directory}/images/Set 2 - Components study.png", bbox_inches='tight')
show()
```

```python
from matplotlib.pyplot import plot, legend

# figure(figsize=(3 * HEIGHT, HEIGHT))
figure()
plot_line_chart(
    series.index.to_list(),
    series.to_list(),
    xlabel=series.index.name,
    ylabel=target,
    title=f"{file_tag} stationary study",
    name="original",
)
n: int = len(series)
plot(series.index, [series.mean()] * n, "r-", label="mean")
legend()

# savefig(f"{path_to_directory}/images/Set 2 - Stationarity study.png", bbox_inches='tight')
show()
```

```python
BINS = 10
mean_line: list[float] = []

for i in range(BINS):
    segment: Series = series[i * n // BINS : (i + 1) * n // BINS]
    mean_value: list[float] = [segment.mean()] * (n // BINS)
    mean_line += mean_value
mean_line += [mean_line[-1]] * (n - len(mean_line))

# figure(figsize=(3 * HEIGHT, HEIGHT))
figure()
plot_line_chart(
    series.index.to_list(),
    series.to_list(),
    xlabel=series.index.name,
    ylabel=target,
    title=f"{file_tag} stationary study",
    name="original",
    show_stdev=True,
)
n: int = len(series)
plot(series.index, mean_line, "r-", label="mean")
legend()

savefig(f"{path_to_directory}/images/Set 2 - Stationarity study.png", bbox_inches='tight')
show()
```

### Augmented Dickey-Fuller test


```python
from statsmodels.tsa.stattools import adfuller


def eval_stationarity(series: Series) -> bool:
    result = adfuller(series)
    print(f"ADF Statistic: {result[0]:.3f}")
    print(f"p-value: {result[1]:.3f}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"\t{key}: {value:.3f}")
    return result[1] <= 0.05


print(f"The series {('is' if eval_stationarity(series) else 'is not')} stationary")
```
